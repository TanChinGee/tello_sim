```python
# install the library from cloned source using the code below
## git clone https://github.com/Fireline-Science/tello_sim
## cd tello_sim
## pip install .

import tello_sim
```

```python
drone = tello_sim.Simulator()
```

```python
# Drone also start with take off
drone.takeoff()
```

```python
# Running thru the entire set of drone commands along the x,y axis
drone.forward(100)
drone.left(50)
drone.back(50)
drone.right(50)
```

```python
# Running thru the entire set of drone commands along the z axis
drone.down(50)
drone.up(50)
```

```python
# Running thru the entire set of drone commands for orientation of yew
drone.cw(180)
drone.ccw(180)
```

```python
# Drone sequence ends with land
drone.land()
```

```python
# Two ways of controlling the real drone
# Method 1: save commands as a json file to be run on the drone
drone.save("my_first_flight.json")
# The code below load the json file to a drone object in the simulation
## my_drone.load_commands(file_path='new_commands.json')

# Method 2: wifi connect to the drone and deploy the commands
drone.deploy()
```
